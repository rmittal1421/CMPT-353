```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy import stats
```

```python
data = pd.read_csv('dog_rates_tweets.csv',  parse_dates=[1])
```

```python
data['rating'] = pd.to_numeric(data['text'].str.extract(pat = r'(\d+(\.\d+)?)/10')[0])
data = data[data.rating < 25]
```

```python
def to_timestamp (date):
    return date.timestamp()
```

```python
data['timestamp'] = data['created_at'].apply(to_timestamp)
fit = linregress(data['timestamp'].values, data['rating'].values)
data['prediction'] = data['timestamp']*fit.slope + fit.intercept
data
```

```python
# fit.slope, fit.intercept
print(stats.linregress(data['timestamp'], data['rating']).rvalue)
```

```python
plt.xticks(rotation=25)
plt.plot(data['created_at'].values, data['rating'].values, 'b.', alpha=0.5)
plt.plot(data['created_at'].values, data['prediction'].values, 'r-', linewidth=3)
plt.xlabel('Date noted on a tweet')
plt.ylabel('Dog rating')
plt.legend(['Actual dog rating observations', 'Best fit line'])
plt.show()
```

```python
print(fit.pvalue)
```

```python
residual = data['rating'] - data['prediction']
plt.hist(residual)
plt.xlabel('Difference b/w onserved & predicted rating')
plt.ylabel('Frequency')
```

```python
plt.xticks(rotation=25)
plt.plot(residual, data['rating'].values, 'b.', alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Residual')
plt.show()
```
