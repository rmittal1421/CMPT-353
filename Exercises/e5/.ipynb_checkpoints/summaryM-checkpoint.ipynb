```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def data_info(filename):
    df = pd.read_csv(filename, dtype=np.float32)
    print(filename)
    print("\tmean:", df.mean().values)
    print("\tstandard deviation:", df.std().values)
    print("\trange = from", df.min().values, "to",df.max().values)
    print("\tcorrelation coefficient (r):", df.x.corr(df.y))
    plt.plot(df.x, df.y, 'r.')
    plt.show()
    print('\n')
```

```python
for i in range(1,7):
    data_info('data-'+str(i)+'.csv')
```
